```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

Sample Data

```python
mu, sigma = 100, 15 # nilai mean dan nilai standard deviation

x = mu + sigma * np.random.randn(10000) #normal distribution
x
```

```python
x.shape
```

```python
plt.hist(x,bins=50, #membentuk 50 kelompok data member
         facecolor='g',#warna grafik
         alpha=0.75) #nilai alpha 0.75 = 75%, 0.25 = 25% dan seterusnya

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Histogram')

plt.text(45,500, r'$\mu=100,\ \sigma=15$')#koordinat memunculkan text(x,y,string yang akan dibuat, r itu optional)
plt.grid() #memunculkan grid

plt.show()
```

Histogram dengan OO Style

```python
fig, ax=plt.subplots()

ax.hist(x,
        bins=50,
        facecolor='r',
        alpha=0.75)

ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')
ax.set_title('Contoh Histogram')

ax.text(45, 500, r'$\mu=100,\ \sigma=15$')
ax.grid()

plt.show()
```
